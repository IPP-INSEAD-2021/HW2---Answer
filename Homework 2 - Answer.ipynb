```python
# character_dict={} (10 Points)

######################################################################################################
# write your code here: 

import pandas as pd

#read the data file
f = open("data/Friends_Transcript.txt", "r")

#set up empty dictionary, and also the list to identify multiple characters
character_dict={}
multi_chara = ['/', ',', 'and', '&', " by", " By"] #Note: " by", " By" are used to eliminate "written by"

for line in f:
    #check whether the first letter of each line with ":" in the script is capitalized or not
    if ":" in line:
        if line[0].isupper() == True: #Note: this also eliminates lines start with "(", "{", "[", " ", "-"
            #split the line and store the characters before ":" as the name
            name = line.split(":")[0]

            #remove lines with multiple characters
            if any(x in name for x in multi_chara):
                continue
            
            else:
                #count the numbers and store as values
                if name in character_dict.keys():
                    character_dict[name] += 1
                else:
                    character_dict[name] = 1

######################################################################################################
# Expected output
character_dict
```

### Problem 8 (20 Points). Using the dataframe you created in Problem 6 that has Region column, create a dictionary called 'region_dict' where the keys are the name of the region and the value is the average happiness score in each year. The format of the dictionary should be the following: 
* {region1={2016:average of score in year 2016 in region1, 2017:average of score in year 2017 in region1, 2018:average of score in year 2018 in region1, 2019:average of score in year 2019 in region1},region2={2016:average,2017:average,2018:average,2019:average}}

```python
# whr_df (20 Points)
# region_dict={}
######################################################################################################
# write your code here: 
Regions=whr_df.Region.unique()
region_dict={}
for region in Regions:
    region_dict[region]={}
    for year in range(2016,2020):
        yearly_score=whr_df[(whr_df['Region']==region) & (whr_df['Year']==year)].Score.mean()
        region_dict[region][year]=yearly_score
######################################################################################################
# Expected output
region_dict
```
